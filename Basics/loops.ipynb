```python
# while loop
i = 1
while i <= 10:
    print(i)
    i += 1
```

```python
# Even numbers from 20 to 40
num = 20
while num <= 40:
    if num % 2 == 0:
        print(num)
    num += 1
```

```python
numbers = [1, 2, 3, 4, 5]
total = 0
for num in numbers:
    print(num)
    total += num
print("Last number:", num)
print("Total:", total)
```

```python
product = 1
for num in numbers:
    product *= num
print("Product:", product)
```
