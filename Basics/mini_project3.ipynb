```python
# Todo list
def todo_manager(todo_list, *tasks, **actions):
    for task in tasks:
        todo_list.append(task)
        print(f"Added task: {task}")

    if actions.get('remove'):
        for task in actions['remove']:
            todo_list[:] = [t for t in todo_list if t != task]
            print(f"Removed task: {task}")

    if actions.get('complete'):
        for task in todo_list:
            if task[0] in actions['complete']:
                task[1] = "done"
                print(f"Completed task: {task[0]}")

    print("\nCurrent Tasks:")
    for task, status in todo_list:
        print(f"- {task} ({status})")

my_tasks = []
todo_manager(my_tasks, "Learn Python", "Lean Python OOPs", "Learn Django", "Play Games", complete=["Learn Python"], remove=["Play Games"])
```
